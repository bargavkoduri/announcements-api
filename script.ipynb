```python
import json
import pymongo
import os
from dotenv import load_dotenv
from dateutil.parser import parse
```

```python
def readFile(fileName):
    with open(fileName, 'r') as file:
        data = json.load(file)
    return data

def convertToDate(data):
    for i in range(len(data)):
        data[i]["NEWS_DT"] = parse(data[i]["NEWS_DT"])

load_dotenv()
MONGO_URL = os.getenv('MONGO_URL')
DB_NAME = os.getenv('DB_NAME')
```

```python
data = readFile('bse_announcements_data.json')
convertToDate(data)
```

```python
client = pymongo.MongoClient(MONGO_URL)
```

```python
db = client[DB_NAME]
```

```python
db.Announcements.insert_many(data)
```
